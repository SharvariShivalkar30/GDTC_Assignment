```python
# Task-6
# Latest date in data
claim_deviation_df = view_df.copy()
claim_deviation_df.TXN_DATE_TIME.max()
```

```python
claim_deviation_df['TXN_DATE_TIME'] = pd.to_datetime(claim_deviation_df['TXN_DATE_TIME'])
Current_date = pd.to_datetime('2021-06-30')
```

```python
Last_30_Days = claim_deviation_df.loc[(claim_deviation_df['TXN_DATE_TIME'] > '2021-05-30') & (claim_deviation_df['TXN_DATE_TIME'] < '2021-06-30')]
```

```python
avg_claim = Last_30_Days.groupby('INSURANCE_TYPE').mean()['CLAIM_AMOUNT'].to_dict()
```

```python
def claim_amnt_deviation(row):
    if (Current_date - row['TXN_DATE_TIME'])<pd.Timedelta(days=30):
        return -1
    claim_amnt_deviation = avg_claim[row['INSURANCE_TYPE']] / row['CLAIM_AMOUNT']
    if claim_amnt_deviation < 0.5:
        return 1
    else:
        return 0
```

```python
claim_deviation_df['claim_amnt_deviation'] = claim_deviation_df.apply(claim_amnt_deviation,axis=1)
```

```python
# Task-8
all_ins_prem = view_df.copy()
total_ins_prem = all_ins_prem.groupby('INSURANCE_TYPE')['PREMIUM_AMOUNT'].sum().to_frame()
total_ins_prem = total_ins_prem.reset_index()
total_ins_prem.loc[(total_ins_prem['INSURANCE_TYPE'] == 'Mobile'), 'Discount by %'] = total_ins_prem['PREMIUM_AMOUNT']*(10/100)
total_ins_prem.loc[(total_ins_prem['INSURANCE_TYPE'] == 'Travel'), 'Discount by %'] = total_ins_prem['PREMIUM_AMOUNT']*(10/100)

total_ins_prem.loc[(total_ins_prem['INSURANCE_TYPE'] == 'Health'), 'Increased by %'] = total_ins_prem['PREMIUM_AMOUNT']*(7/100)
total_ins_prem.loc[(total_ins_prem['INSURANCE_TYPE'] == 'Property'), 'Increased by %'] = total_ins_prem['PREMIUM_AMOUNT']*(7/100)
total_ins_prem.loc[(total_ins_prem['INSURANCE_TYPE'] == 'Life'), 'Increased by %'] = total_ins_prem['PREMIUM_AMOUNT']*(2/100)
total_ins_prem.loc[(total_ins_prem['INSURANCE_TYPE'] == 'Motor'), 'Increased by %'] = total_ins_prem['PREMIUM_AMOUNT']*(2/100)

total_ins_prem.loc[(total_ins_prem['INSURANCE_TYPE'] == 'Mobile') | (total_ins_prem['INSURANCE_TYPE'] == 'Travel'),
                   'After Discount or Increased'] = total_ins_prem['PREMIUM_AMOUNT'] - total_ins_prem['Discount by %']

total_ins_prem.loc[(total_ins_prem['INSURANCE_TYPE'] == 'Health') | (total_ins_prem['INSURANCE_TYPE'] == 'Property') |
                   (total_ins_prem['INSURANCE_TYPE'] == 'Life') | (total_ins_prem['INSURANCE_TYPE'] == 'Motor'),
                   'After Discount or Increased'] = total_ins_prem['PREMIUM_AMOUNT'] + total_ins_prem['Increased by %']
total_ins_prem['After Discount or Increased'] = total_ins_prem['After Discount or Increased'].round(decimals=2)
overall_prem_amnt = total_ins_prem['PREMIUM_AMOUNT'].sum()
overall_discount = total_ins_prem['After Discount or Increased'].sum()
percentage_diff = ((overall_discount - overall_prem_amnt) / overall_prem_amnt) * 100
percentage_diff = percentage_diff.round(decimals=2)
print(total_ins_prem.to_string())
print("\nOverall Premium Amount = ", overall_prem_amnt, "\nOverall Discount Amount = ", overall_discount,
      "\nOverall Percentage Change of the Premium Amount Collected = ", percentage_diff, "%")
```
